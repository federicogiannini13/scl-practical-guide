```python
from sklearn.datasets import fetch_openml

# Scarica il dataset MNIST
mnist = fetch_openml('mnist_784', version=1)
```

```python
df = mnist["data"]
```

```python
df["target"] = mnist["target"]
```

```python
df["target"] = df["target"].astype(int)
```

```python
df["task1"] = df["target"].apply(lambda x : d["even"][x])
df["task2"] = df["target"].apply(lambda x : d["greater_4"][x])
df["task3"] = df["target"].apply(lambda x : d["multiple_3"][x])
df["task4"] = df["target"].apply(lambda x : d["prime"][x])
df["task5"] = df["target"].apply(lambda x : d["not_outlier"][x])
```

```python
for i in range(1, 785):
    df[f"pixel{i}"] = df[f"pixel{i}"] / 255
```

```python
df = df.sample(frac=1).reset_index(drop=True)
```

```python
train_df, test_df = train_test_split(df, test_size=0.1667, stratify=df['target'])
train_df = train_df.reset_index(drop=True)
test_df = test_df.reset_index(drop=True)
```

```python
from sklearn.model_selection import train_test_split
import os
import pandas as pd
from sklearn.manifold import TSNE
import umap
```

```python
root = "/Users/federicogiannini/Library/CloudStorage/OneDrive-PolitecnicodiMilano/SML_CL"

df = pd.read_csv(os.path.join(root, "datasets", "mnist.csv"))
train_df = pd.read_csv(os.path.join(root, "datasets", "mnist_train.csv"))
test_df = pd.read_csv(os.path.join(root, "datasets", "mnist_test.csv"))
```

```python
umap_model = umap.UMAP(n_components=30)
umap_model.fit(train_df.iloc[:,:784].values)
```

```python
X_train_red = umap_model.transform(train_df.iloc[:,:784].values)
X_test_red = umap_model.transform(test_df.iloc[:,:784].values)
```

```python
def red(X, df):
    df_red = pd.DataFrame(X, columns=[f"feat{i}" for i in range(X.shape[1])])
    df_red["target"] = list(df["target"])
    for i in range(1,6):
        df_red[f"task{i}"] = list(df[f"task{i}"])
    return df_red
```

```python
df_train_red = red(X_train_red, train_df)
df_test_red = red(X_test_red, test_df)
```

```python
def make_exp_incremental(df):
    dfs = []
    experiences = [(i,i+1) for i in range(0,10,2)]
    for exp in experiences:
        df_exp = df[df["target"].isin(exp)].reset_index(drop=True)
        df_exp["target"] = df_exp["task1"]
        df_exp = df_exp.drop(columns=[f"task{i}" for i in range(1,6)]).sample(frac=1).reset_index(drop=True)
        dfs.append(df_exp)
    return dfs
```

```python
exps_train = make_exp_incremental(df_train_red)
exps_test = make_exp_incremental(df_test_red)
```

```python
for i in range(len(exps_train)):
    exps_train[i].to_csv(os.path.join(root, "datasets", f"mnist_red30_incremental_exp{i+1}_train.csv"), index=False)
    exps_test[i].to_csv(os.path.join(root, "datasets", f"mnist_red30_incremental_exp{i+1}_test.csv"), index=False)
for i in range(len(exps_train)):
    exps_train[i]["task"] = i+1
    exps_test[i]["task"] = i+1
pd.concat(exps_train).reset_index(drop=True).to_csv(os.path.join(root, "datasets", "mnist_red30_incremental_train.csv"), index=False)
pd.concat(exps_test).reset_index(drop=True).to_csv(os.path.join(root, "datasets", "mnist_red30_incremental_test.csv"), index=False)
```

```python
def make_exp_sml(df):
    df_ = df.copy()
    n_split = 5
    samples_per_split = len(df_) // n_split
    dfs = []

    for target_value, group in df_.groupby('target'):
        for i in range(n_split):
            sampled_group = group.sample(n=min(samples_per_split, len(group)))
            dfs.append(sampled_group)
    final_dfs = [pd.concat(dfs[i::n_split]).reset_index(drop=True) for i in range(n_split)]
    for i in range(len(final_dfs)):
        final_dfs[i]["target"] = final_dfs[i][f"task{i+1}"]
        final_dfs[i] = final_dfs[i].drop(columns=[f"task{i}" for i in range(1,6)])
    return final_dfs
```

```python
exps_sml_train = make_exp_sml(df_train_red)
exps_sml_test = make_exp_sml(df_test_red)
```

```python
for i in range(len(exps_train)):
    exps_sml_train[i].to_csv(os.path.join(root, "datasets", f"mnist_red30_sml_exp{i+1}_train.csv"), index=False)
    exps_sml_test[i].to_csv(os.path.join(root, "datasets", f"mnist_red30_sml_exp{i+1}_test.csv"), index=False)
for i in range(len(exps_sml_train)):
    exps_sml_train[i]["task"] = i+1
    exps_sml_test[i]["task"] = i+1
pd.concat(exps_sml_train).reset_index(drop=True).to_csv(os.path.join(root, "datasets", "mnist_red30_sml_train.csv"), index=False)
pd.concat(exps_sml_test).reset_index(drop=True).to_csv(os.path.join(root, "datasets", "mnist_red30_sml_test.csv"), index=False)
```
